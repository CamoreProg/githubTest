# Connect to the Gate io API

#### Connection setting

```python
import requests

url = 'https://api.gateio.ws/api/v4'
#testing with 
testapi = '/spot/tickers'
#headers = {'Authorization': 'Bearer your_access_token'}
#params = {'param1': 'value1', 'param2': 'value2'}

#info: https://github.com/gateio/gateapi-python
```

#### Test data connection

```python
response = requests.get(url+testapi)

if response.status_code == 200:
    data = response.json()
    print('Success!')
    print(data)
else:
    print('Error: ', response.status_code)
    
```

#### Test retrieving Aleph Zero pair

```python
icurrency_pairs = ['BTC_USDT', 'KAS_USDT', 'MNW_USDT', 'ARB_USDT', 'ETH_USDT']

for icurrency_pair in icurrency_pairs:
    
    for innerlist in data:
        currency_pair = innerlist['currency_pair']
        if currency_pair == icurrency_pair:
            print('The current value of ' + currency_pair + ' is ' + innerlist.get('last'))
            
        
```
